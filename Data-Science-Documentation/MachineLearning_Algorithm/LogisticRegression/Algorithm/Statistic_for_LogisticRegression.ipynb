###(1) $h^{-1}(p)=\ln(\frac{p}{1-p})$

**Prove that $h(p)=\frac{1}{1 + \exp({-\vec{h}\cdot\vec{x}})},h^{-1}(p)=\ln(\frac{p}{1-p})$**

**What is Odds?**

Odds provide a measure of the likelihood of a particular outcome.

* Odds $=\frac{\text{probability that the outcome occurs}}{\text{the probability that the outcome does not occur}}=\frac{p}{1-p}$
* Odds is not probability
* Probability $=\frac{\text{probability that the outcome occurs}}{\text{total outcome}}=p\in[0,1]$

**What is $\ln(\frac{p}{1-p})$ used for?**

* $\ln()$ makes odds becomes symmetrical.
* $\ln(\frac{1}{a})-\ln(a)=0, a\in\mathbb{R}$
* $h^{-1}(p)=\ln(\frac{p}{1-p})$

**Prove that $\ln(\frac{1}{a})-\ln(a)=0$**

According to Logarithm: $\log_a(\frac{x}{y})=\log_a(x)-\log_a(y)$

$\therefore\ln(\frac{1}{a})-\ln(a)=\ln(1)=\log_e(1)$

According to Logarithm: $a^b=c\implies\log_a(c)=b$

$\therefore e^0=1\implies\log_e(1)=0$

$\frac{d}{dx}[\frac{1}{e^{m*x}}]=\frac{d}{dx}[e^{-mx}]$

Given that:
* $f(x)=e^x$
* $g(x)=-mx$

According to derivative of $e$: $\frac{d}{dx}[e^x] =e^x$:
* $\therefore\frac{d}{dx}f(x)=e^x$

According to Power Rule: $f'(x^n) = nx^{n-1}$

According to Constant multiple rule No.1, $\frac{d}{dx}[k*f(x)] = k*\frac{d}{dx}f(x)$
* $\therefore \frac{d}{dx}g(x)=-m$

$\therefore\frac{d}{dx}[\frac{1}{e^x}]=e^{-x}*-m$

$\therefore\frac{d}{dx}[\frac{1}{e^x}]=-me^{-x}$

$\frac{d}{dx}[\frac{1}{e^{m*x+c}}]=\frac{d}{dx}[e^{-mx}*e^c]$

According to Constant multiple rule No.1, $\frac{d}{dx}[k*f(x)] = k*\frac{d}{dx}f(x)$

$\frac{d}{dx}[\frac{1}{e^{m*x+c}}]=e^c*\frac{d}{dx}[e^{-mx}]$

$\therefore\frac{d}{dx}[\frac{1}{e^{m*x+c}}]=-m*\frac{e^c}{e^x}$
