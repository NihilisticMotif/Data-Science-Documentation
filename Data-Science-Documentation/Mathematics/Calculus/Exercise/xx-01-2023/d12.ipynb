###(2) $(x+y)^{x+y}=?$

**Find $(x+y)^{x+y}=?$**

According to Binomial theorem: $(x+y)^n=\Sigma_{r=0}^n\frac{n!}{(n-r)!*r!}x^{n-r}y^r$

$\therefore(x+y)^{x+y}=\Sigma_{r=0}^{x+y}\frac{(x+y)!}{((x+y)-r)!*r!}x^{x+y-r}y^r$

###(3) $\frac{d}{dx}[x^x]$

**Find $\frac{d}{dx}[x^x]$ Solution No.1**

$y=x^x$

$\ln(y)=\ln(x^x)$

According to Logarithm: $\log_a(x^y)=y*\log_a(x)=\log_a(x)*y$

$\therefore\ln(y)=x*\ln(x)$

$\frac{dy}{dx}[\ln(y)]=\frac{dy}{dx}[x*\ln(x)]$

According to derivative: $\frac{d}{dx}[\ln(x)]=\frac{1}{x}$

$\therefore\frac{dy}{dx}*\frac{1}{y}=\frac{dy}{dx}[x*\ln(x)]$

According to Product Rule, $\frac{d}{dx}[f(x)*g(x)] = f'(x)*g(x)+g'(x)*f(x)$

$\therefore\frac{dy}{dx}*\frac{1}{y}=\frac{dy}{dx}[x]*\ln(x)+\frac{dy}{dx}[\ln(x)]*x$

$\frac{dy}{dx}*\frac{1}{y}=\ln(x)+\frac{1}{x}*x$

$\frac{dy}{dx}*\frac{1}{y}=\ln(x)+1$

$\frac{dy}{dx}=y(\ln(x)+1)$

Given that: $y=x^x$

$\therefore\frac{dy}{dx}=x^x(\ln(x)+1)$

**Find $\frac{d}{dx}[x^x]$ Solution No.2**

$\frac{d}{dx}[x^x]=\lim_{h\to0}\frac{(x+h)^{x+h}-x^x}{h}$

According to Binomial theorem: $(x+y)^n=\Sigma_{r=0}^n\frac{n!}{(n-r)!*r!}x^{n-r}y^r$

$\therefore\frac{d}{dx}[x^x]=\lim_{h\to0}\frac{1}{h}*(\Sigma_{r=0}^{x+h}\frac{(x+h)!}{(x+h-r)!*r!}x^{x+h-r}h^r-x^x)$

$\frac{d}{dx}[x^x]=\lim_{h\to0}\frac{1}{h}*(x^{x+h}+(x+h)x^{x+h-1}h+\frac{(x+h)(x+h-1)}{2}x^{x+h-2}h^2+...-x^x)$

$\frac{d}{dx}[x^x]=\lim_{h\to0}\frac{1}{h}*(x^{x+h}-x^x+(x+h)x^{x+h-1}h+\frac{(x+h)(x+h-1)}{2}x^{x+h-2}h^2+...)$

$\frac{d}{dx}[x^x]=\lim_{h\to0}\frac{x^{x+h}-x^x}{h}+\lim_{h\to0}\frac{1}{h}((x+h)x^{x+h-1}h+\frac{(x+h)(x+h-1)}{2}x^{x+h-2}h^2+...)$

$\frac{d}{dx}[x^x]=x^x*\lim_{h\to0}\frac{x^{h}-1}{h}+\lim_{h\to0}\frac{1}{h}((x+h)x^{x+h-1}h+\frac{(x+h)(x+h-1)}{2}x^{x+h-2}h^2+...)$

Given that: $\lim_{h\to0}\frac{x^{h}-1}{h}=\ln(x)$

$\frac{d}{dx}[x^x]=x^x*\ln(x)+\lim_{h\to0}\frac{1}{h}((x+h)x^{x+h-1}h+\frac{(x+h)(x+h-1)}{2}x^{x+h-2}h^2+...)$

$\frac{d}{dx}[x^x]=x^x*\ln(x)+\lim_{h\to0}((x+h)x^{x+h-1}+\frac{(x+h)(x+h-1)}{2}x^{x+h-2}h+...)$

$\frac{d}{dx}[x^x]=x^x*\ln(x)+\lim_{h\to0}((x+h)x^{x+h-1}+h*(\frac{(x+h)(x+h-1)}{2}x^{x+h-2}+...))$

$\frac{d}{dx}[x^x]=x^x*\ln(x)+\lim_{h\to0}((x+h)x^{x+h-1})$

$\frac{d}{dx}[x^x]=x^x*\ln(x)+((x)x^{x-1})$

$\therefore\frac{d}{dx}[x^x]=x^x*\ln(x)+x^x$

###(6) $\lim_{h \to 0}\frac{e^{h}-1}{h}=1$ and $\lim_{h \to 0}\frac{c^{h}-1}{h}=\ln(c)$

**Proof No.1: $\lim_{h \to 0}\frac{e^{h}-1}{h}=1$**

Given that
* $y=e^h-1$

$\therefore y+1=e^h$

$\log_e(y+1)=\log_e(e^h)$

According to Logarithm property: $\log_aa^x=x$

$\therefore \log_e(y+1)=h$

Given that
* $y=e^h-1$
* $\log_e(y+1)=h$

$\therefore \frac{e^{h}-1}{h}=\frac{y}{\log_e(y+1)}=\frac{e^{h}-1}{\log_e(e^h-1+1)}$

Given that:
* $\lim_{h\to0}e^h-1=\lim_{y\to0}y$

$\therefore \lim_{h\to0}\frac{e^{h}-1}{h}=\lim_{y\to0}\frac{y}{\log_e(y+1)}$

$\lim_{y\to0}\frac{y}{\ln(y+1)}=\lim_{y\to0}\frac{1}{\frac{\ln(y+1)}{y}}$

$\lim_{y\to0}\frac{y}{\ln(y+1)}=\lim_{y\to0}\frac{1}{\frac{1}{y}*\ln(y+1)}$


According to Logarithm property: $\log_a(x^y)=y*\log_a(x)=\log_a(x)*y$

$\therefore\lim_{y\to0}\frac{y}{\ln(y+1)}=\lim_{y\to0}\frac{1}{\ln((y+1)^\frac{1}{y})}$

Given that
* $\lim_{n\to\infty}(1+\frac{1}{n})^n=e$
* $\lim_{n\to\infty}\frac{1}{n}=\lim_{y\to0}y$
* $\therefore \lim_{n\to\infty}(1+\frac{1}{n})^n=\lim_{y\to0}(y+1)^\frac{1}{y}=e$

$\therefore\lim_{y\to0}\frac{y}{\ln(y+1)}=\frac{1}{\ln(e)}$

According to Logarithm property: $\log_aa^x=x$

$\therefore\lim_{y\to0}\frac{y}{\ln(y+1)}=\frac{1}{1}=1$

**Proof No.2: $\lim_{h\to0}\frac{c^{h}-1}{h}=\log_e(c)=\ln(c)$**

Given that:
* $y=c^h-1$

$y+1=c^h$

$\ln(y+1)=\ln(c^h)$

According to Logarithm: $\log_a(x^y)=y*\log_a(x)=\log_a(x)*y$

$\therefore\ln(y+1)=h*\ln(c)$

$\frac{\ln(y+1)}{\ln(c)}=h$

Given that:
* $y=c^h-1$
* $\frac{\ln(y+1)}{\ln(c)}=h$

$\therefore \frac{c^h-1}{h}=\frac{y}{\frac{\ln(y+1)}{\ln(c)}}$

$\frac{c^h-1}{h}=\frac{y}{\ln(y+1)}*\ln(c)$
