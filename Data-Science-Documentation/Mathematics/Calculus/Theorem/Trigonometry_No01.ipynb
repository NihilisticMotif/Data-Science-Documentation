###(6) $\frac{d}{dx}[\sin(x)]=\cos(x)$ and $\frac{d}{dx}[\cos(x)]=-\sin(x)$

Prove that $\frac{d}{dx}[\sin(x)]=\cos(x)$:

$\frac{d}{dx}[\sin(x)]=\lim_{h \to 0}\frac{\sin(x+h)-sin(x)}{h}$

$=\lim_{h \to 0}\frac{\sin(x)\cos(h)+\cos(x)\sin(h)-sin(x)}{h}$

$=\lim_{h\to0}\frac{\sin(x)\cos(h)-\sin(x)}{h}+ \lim_{h\to0}\frac{\cos(x)*\sin(h)}{h}$

$=\sin(x)*\lim_{h\to0}\frac{\cos(h)-1}{h}+ \cos(x)*\lim_{h\to0}\frac{\sin(h)}{h}$

$=\sin(x)*0+ \cos(x)*1$

$\therefore \frac{d}{dx}[\sin(x)]=\cos(x)$

Prove that $\frac{d}{dx}[\cos(x)]=-\sin(x)$:

$\frac{d}{dx}[\cos(x)]=\lim_{h\to0}\frac{\cos(x+h)-\cos(x)}{h}$

$=\lim_{h\to0}\frac{\cos(x)\cos(h)-\sin(x)\sin(h)-\cos(x)}{h}$

$=\lim_{h\to0}\frac{\cos(x)\cos(h)-\cos(x)-\sin(x)\sin(h)}{h}$

$=\lim_{h\to0}\frac{\cos(x)\cos(h)-\cos(x)}{h} + \lim_{h\to0}\frac{-\sin(x)\sin(h)}{h}$

$=\cos(x)*\lim_{h\to0}\frac{\cos(h)-1}{h} + -\sin(x)*\lim_{h\to0}\frac{\sin(h)}{h}$

$=\cos(x)*0 + -\sin(x)*1$

$\therefore \frac{d}{dx}[\cos(x)]=-\sin(x)$

###(7) $\frac{d}{dx}[\tan(x)]=\frac{1}{\cos(x)^2}$

According to trigometry: $\tan(x)=\frac{\sin(x)}{\cos(x)}$

$\therefore\frac{d}{dx}[\tan(x)]=\frac{d}{dx}[\frac{\sin(x)}{\cos(x)}]$

$\therefore\frac{d}{dx}[\tan(x)]=\frac{d}{dx}[\sin(x)*\cos(x)^{-1}]$

According to Product Rule: $\frac{d}{dx}[f(x)*g(x)] = f'(x)*g(x)+g'(x)*f(x)$

$\frac{d}{dx}[\tan(x)]=\frac{d}{dx}[\sin(x)]*\cos(x)^{-1}+\sin(x)*\frac{d}{dx}[\cos(x)^{-1}]$

According to Trigonometry: $\frac{d}{dx}[\sin(x)]=\cos(x)$

$\therefore\frac{d}{dx}[\tan(x)]=\cos(x)*\cos(x)^{-1}+\sin(x)*\frac{d}{dx}[\cos(x)^{-1}]$

According to Chain Rule: $\frac{d}{dx}[f(g(x))] = (f\circ g)'= f'(g(x))*g'(x)$

Given that:
* $f(x)=x^{-1}\implies f'(x)=-x^{-2}=\frac{-1}{x^2}$
* $g(x)=\cos(x)\implies g'(x)=-\sin(x)$

$\therefore\frac{d}{dx}[\tan(x)]=\cos(x)*\cos(x)^{-1}+\sin(x)*
-\sin(x)*\frac{-1}{\cos(x)^2}$

$\frac{d}{dx}[\tan(x)]=1-\sin(x)*\sin(x)*\frac{-1}{\cos(x)^2}$

$\frac{d}{dx}[\tan(x)]=1+\sin(x)*\sin(x)*\frac{1}{\cos(x)^2}$

$\frac{d}{dx}[\tan(x)]=1+\frac{\sin(x)^2}{\cos(x)^2}$

$\frac{d}{dx}[\tan(x)]=\frac{\cos(x)^2+\sin(x)^2}{\cos(x)^2}$

According to trigonometry: $\cos(x)^2+\sin(x)^2=1$

$\therefore\frac{d}{dx}[\tan(x)]=\frac{1}{\cos(x)^2}$

###(8) $\lim_{x\to 0 }\frac{\cos(x)-1}{x}=0$

**Prove that $\lim_{x\to 0 }\frac{\cos(x)-1}{x}=0$:**

$\lim_{x\to 0 }\frac{\cos(x)-1}{x}=\lim_{x\to 0 }\frac{\cos(x)-1}{x}*\frac{\cos(x)+1}{\cos(x)+1}$

$=\lim_{x\to 0 }\frac{\cos(x)^2-1}{x(\cos(x)+1)}$

According to Pythagorean Identities: $1=\sin(x)^2+cos(x)^2$

$=\lim_{x\to 0 }\frac{\cos(x)^2-\sin(x)^2-cos(x)^2}{x(\cos(x)+1)}$

$=\lim_{x\to 0 }\frac{-\sin(x)^2}{x(\cos(x)+1)}$

According to Trigonometry: $1+\cos(x)=2\cos(\frac{x}{2})^2$

$=\lim_{x\to 0 }\frac{-\sin(x)^2}{x(2\cos(\frac{x}{2})^2)}$

According to Trigonometry: $\sin(x)=2\sin(\frac{x}{2})\cos(\frac{x}{s})$

$=\lim_{x\to 0 }\frac{-4\sin(\frac{x}{2})^2\cos(\frac{x}{2})^2}{x(2\cos(\frac{x}{2})^2)}$

$=\lim_{x\to 0 }\frac{-2\sin(\frac{x}{2})^2}{x}$

According to LHospitals Rule, $\lim_{x \to a}\frac{f(x)}{g(x)} = \lim_{x \to a}\frac{f'(x)}{g'(x)} $

$=\lim_{x\to 0 }\frac{\frac{d}{dx}[-2\sin(\frac{x}{2})^2]}{\frac{d}{dx}[x]}$

$=\lim_{x\to 0 }\frac{d}{dx}[-2\sin(\frac{x}{2})^2]$

According to Chain Rule: $\frac{d}{dx}[f(g(x))] = (f\circ g)'= f'(g(x))*g'(x)$

Given: $f(x)=-2x^2 \implies f'(x)=-4x$

Given: $g(x)=\sin(\frac{x}{2})$

$\frac{d}{dx}[f(g(x))]=f'(g(x))*g'(x)$

Given that: 
* $g_1(x)=\sin(x) \implies g_1'(x)=\cos(x)$
* $g_2(x)=\frac{x}{2} \implies g_2'(x)=\frac{1}{2}$
* $g(x)=g_1(g_2(x)) \implies \frac{d}{dx}[g(x)]=g_1'(g_2(x))*g_2'(x)$
* $\frac{d}{dx}[f(g(x))]=f'(g(x))*g'(x)$
* $\frac{d}{dx}[f(g(x))]=f'(g_1(g_2(x)))*g_1'(g_2(x))*g_2'(x)$

In this case:

$\lim_{x\to 0 }\frac{d}{dx}[-2\sin(\frac{x}{2})^2]=\lim_{x\to 0 }-4\sin(\frac{x}{2})*\cos(\frac{x}{2})*\frac{1}{2}$

$\lim_{x\to 0 }\frac{d}{dx}[-2\sin(\frac{x}{2})^2]=\lim_{x\to 0 }-2\sin(\frac{x}{2})*\cos(\frac{x}{2})$

According to Trigonometry: $\sin(x)=2\sin(\frac{x}{2})*\cos(\frac{x}{2})$

$\lim_{x\to 0 }\frac{d}{dx}[-2\sin(\frac{x}{2})^2]=\lim_{x\to 0 }-\sin(x)$

$\lim_{x\to 0 }\frac{d}{dx}[-2\sin(\frac{x}{2})^2]=-\sin(0) = 0$

$\therefore \lim_{x\to 0 }\frac{\cos(x)-1}{x}=0$
