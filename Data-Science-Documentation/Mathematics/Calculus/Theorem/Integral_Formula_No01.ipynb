**Q: What is integral used for?**

A: It is used for finding area under the curve and convert derivative function to its original form.

**Type of Integral**
1. definite integral (Area under the curve) $=\int_{a}^{b} f(x) \,dx$
2. indefinite integral (antiderivative) $=\int f(x)\,dx$

**Definite Integral Formula No.1 (Riemann sum)**

* Integral is a way to find area under the curve of any function by define the area of the function as an arrangement of infinite rectangles.
* Every rectangle have same width, but some rectangles have different height.

$\int_{a}^{b} f(x) \,dx=\Sigma f(x_i)*\Delta x=$ area under the curve $f(x)$ at point $a$ to $b$ on the $x$ axis

$\int_{a}^{b} f(x) \,dx=\lim_{n\to\infty}\Sigma_{i=1}^nf(a+\frac{b-a}{n}*i)*(\frac{b-a}{n})=$ area under the curve $f(x)$ at point $a$ to $b$ on the $x$ axis

* $a=$ Starting point on the axis
* $b=$ Ending point on the axis
* $\lim_{n\to\infty}f(a+\frac{b-a}{n}*i)*(\frac{b-a}{n})=$ a very thin rectangle that is also a part of a area under the curve.
* $\lim_{n\to\infty}(\frac{b-a}{n})=\Delta x=$ Width of a very thin rectangle
* $\lim_{n\to\infty}f(a+\frac{b-a}{n}*i)=x_i=$ Height of a very thin rectangle

**Proof of Fundamental theorem of calculus**

According to integral: $\int_{a}^{b} f'(x) \,dx=f(b)-f(a)$

$\therefore \lim_{n\to\infty}f(t+\frac{1}{n})-f(t)=\lim_{n\to\infty}\int_{t}^{t+\frac{1}{n}} f'(x) \,dx$

According to integral: $\int_{a}^{b} f(x) \,dx=\lim_{n\to\infty}\Sigma_{i=1}^nf(a+\frac{b-a}{n}*i)*(\frac{b-a}{n})$

$\therefore\lim_{n\to\infty}f(t+\frac{1}{n})-f(t)=\lim_{n\to\infty}\Sigma_{i=1}^nf(t+\frac{t+\frac{1}{n}-t}{n}*i)*(\frac{t+\frac{1}{n}-t}{n})$

$\lim_{n\to\infty}f(t+\frac{1}{n})-f(t)=\lim_{n\to\infty}\Sigma_{i=1}^nf(t+\frac{1}{n^2}*i)*(\frac{1}{n^2})$

$\lim_{n\to\infty}(f(t+\frac{1}{n})-f(t))*n=\lim_{n\to\infty}\Sigma_{i=1}^nf(t+\frac{1}{n^2}*i)*\frac{1}{n}$

$\lim_{n\to\infty}(f(t+\frac{1}{n})-f(t))*n=\Sigma_{i=1}^nf(t)$

$\therefore\lim_{h\to0}\frac{f(t+h)-f(t)}{h}=\Sigma_{i=1}^nf(t)$

###(2) Integral Multiplication by constant rule: $\int a \,dx=a\int \,dx=a*(x+c)$

**Prove that: $\int a \,dx=a\int \,dx=a*(x+c)$**

According to Constant multiple rule: $\frac{d}{dx}[k*f(x)] = k*\frac{d}{dx}f(x)$

$\frac{d}{dx}[k*(x+c)] = k$

According to integral: $\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$

$\int k\ dx = k*(x+c)$

Given that
* According to integral: $\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$
* According to Power Rule: $f'(x^n) = nx^{n-1}$
* $\therefore \frac{d}{dx}[x]=1\implies \int 1 \ dx=\int \ dx=x+c$

$\therefore \int k\ dx = k\int\ dx = k*(x+c)$

$\frac{d}{dx}[k*x] = k*\frac{d}{dx}f(x)$

###(3) Integral Power Rule $\int x^n\ dx= \frac{x^{n+1}}{n+1}+c|n\neq-1$

**Prove that $\int x^n\ dx= \frac{x^{n+1}}{n+1}+c$:**

According to Power Rule: $f'(x^n) = nx^{n-1}$

$\frac{d}{dx}(x^n) = nx^{n-1}$

Given that
* $g(x) = \frac{d}{dx}(x^n) = nx^{n-1}$
* $f(x) = x^n$

According to integral: $\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$

$\therefore \int nx^{n-1} \ dx=x^n+c$

According to Integral Multiplication by constant rule: $\int a \,dx=a\int \,dx=a*(x+c)$

$\therefore n\int x^{n-1} \ dx=x^n+c$

$\int x^{n-1} \ dx=\frac{x^n+c}{n}$

Given that
* $m=n-1$

$\therefore \int x^m \ dx=\frac{x^{m+1}+c}{m+1}$
