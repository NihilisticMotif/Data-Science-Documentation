###(1) Introduction $\int_{a}^{b} f(x) \,dx$

**Q: What is integral used for?**

A: It is used for finding area under the curve and convert derivative function to its original form.

**What is Integral?**
* Area under curve

$\int_{a}^{b} f(x) \,dx=lim_{n\to\infty}\Sigma_{i=1}^nf(a+\frac{a-b}{n}*i)*(\frac{a-b}{n})=$ area under the curve $f(x)$ at point $a$ to $b$ on the $x$ axis

* Anti-Derivative

$\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$


**Why Integral is defined in this way?**

A: 

Given that
* $x_1$ is starting point
* $x_2$ is ending point

1. Identify starting and ending point ($x_1$ and $x_2$) on x-axis.
2. Split the area under curve into many rectangle as this picture from starting to ending point.
3. calculate area of each rectangles and add all of them

Given that
* $\lim_{q\to x}q-x=\lim_{n\to\infty}\frac{1}{n}=\lim_{h\to0}h=$ width of each rectangle
* $f(x)=$ height of each rectangle

$\therefore \text{area of each rectangle}=\lim_{n\to\infty}\frac{1}{n}*f(x)$

Given that
* $\lim_{n\to\infty}n=$ number of all rectangle.
* $x_1$ is starting point
* $x_2$ is ending point
* $\Delta x=x_2-x_1$

$\therefore \text{area of all rectangle } = \int_{x_1}^{x_2} f(x) \,dx=lim_{n\to\infty}\Sigma_{i=1}^nf(x_1+\frac{\Delta x}{n}*i)*(\frac{\Delta x}{n})$

Note that
* $\text{area of all rectangle }\neq \lim_{n\to\infty}(n+1)*f(x)=\lim_{n\to\infty}\Sigma_{i=0}^nf(x)$
* In other words, $i\text{ in }f(x_1+\frac{\Delta x}{n}*i)$ is used for update new input value of $f(x)$

Why $\Delta x$?


**Q: What is Mesh Size?**

A: When a closed interval $[a,b]$ is partitioned by points $a<x_1<x_2<...<x_{n-1}<b$, the lengths of the resulting intervals between the points are denoted $\Delta x_1, \Delta x_2, ..., \Delta x_n$, and the value max $\Delta x_k$ is called the mesh size of the partition.

###(2) Integral Multiplication by constant rule: $\int a \,dx=a\int \,dx=a*(x+c)$

**Prove that: $\int a \,dx=a\int \,dx=a*(x+c)$**

According to Constant multiple rule: $\frac{d}{dx}[k*f(x)] = k*\frac{d}{dx}f(x)$

$\frac{d}{dx}[k*(x+c)] = k$

According to integral: $\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$

$\int k\ dx = k*(x+c)$

Given that
* According to integral: $\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$
* According to Power Rule: $f'(x^n) = nx^{n-1}$
* $\therefore \frac{d}{dx}[x]=1\implies \int 1 \ dx=\int \ dx=x+c$

$\therefore \int k\ dx = k\int\ dx = k*(x+c)$

**Prove (Unfinished)**

According to Constant multiple rule: $\frac{d}{dx}[k*f(x)] = k*\frac{d}{dx}f(x)$

$\frac{d}{dx}[k*x] = k*\frac{d}{dx}f(x)$

According to integral: $\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$

$\int[k*\frac{d}{dx}f(x)]\ dx = k*f(x)$

###(3) Integral Power Rule $\int x^n\ dx= \frac{x^{n+1}}{n+1}+c|n\neq-1$

**Prove that $\int x^n\ dx= \frac{x^{n+1}}{n+1}+c$:**

According to Power Rule: $f'(x^n) = nx^{n-1}$

$\frac{d}{dx}(x^n) = nx^{n-1}$

Given that
* $g(x) = \frac{d}{dx}(x^n) = nx^{n-1}$
* $f(x) = x^n$

According to integral: $\frac{d}{dx}[f(x)]= g(x) \implies \int g(x) \ dx=f(x)$

$\therefore \int nx^{n-1} \ dx=x^n+c$

According to Integral Multiplication by constant rule: $\int a \,dx=a\int \,dx=a*(x+c)$

$\therefore n\int x^{n-1} \ dx=x^n+c$

$\int x^{n-1} \ dx=\frac{x^n+c}{n}$

Given that
* $m=n-1$

$\therefore \int x^m \ dx=\frac{x^{m+1}+c}{m+1}$
