###(1) Inverse Function of Trigometry function

Note:
* We have to choose any domain and range of a function that satisfy the property of an inverse function

**Function No.1**
* Function: $y=\sin(x)$
* Domain of a function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Range of a function: $[-1,1]$
* Inverse Function: $y=\arcsin(x)$
* Domain of an inverse function: $[-1,1]$
* Range of an inverse function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Note: $\arcsin(x)\neq\frac{1}{\sin(x)}$

**Function No.2**
* Function: $y=\cos(x)$
* Domain of a function: $[0,\pi]$
* Range of a function: $[-1,1]$
* Inverse Function: $y=\arccos(x)$
* Domain of an inverse function: $[-1,1]$
* Range of an inverse function: $[0,\pi]$
* Note: $\arccos(x)\neq\frac{1}{\cos(x)}$

**Function No.1**
* Function: $y=\tan(x)$
* Domain of a function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Range of a function: $[-\infty,\infty]$
* Inverse Function: $y=\arctan(x)$
* Domain of an inverse function: $[-\infty,\infty]$
* Range of an inverse function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Note: $\arctan(x)\neq\frac{1}{\tan(x)}$

###(2) Derivative of an Inverse Function of Trigometry function

* $\frac{d}{dx}[\arcsin(x)]=(1-x^2)^{\frac{-1}{2}}$
* $\frac{d}{dx}[\arccos(x)]=-(1-x^2)^{\frac{-1}{2}}$
* $\frac{d}{dx}[\arctan(x)]=(1+x^2)^{-1}$

**Find: $\frac{d}{dx}[\arcsin(x)]$**

Given that: $y=\arcsin(x)$

$\therefore \sin(y)=x$

$\frac{dy}{dx}[\sin(y)]=\frac{dy}{dx}[x]$

$\frac{dy}{dx}[\sin(y)]=1$

According to Chain Rule: $\frac{d}{dx}[f(g(x))] = (f\circ g)'= f'(g(x))*g'(x)$

According to Trigometry: $\frac{d}{dx}[\sin(x)]=\cos(x)$

Given that:
* $f(x)=\sin(x)\implies f'(x)=\cos(x)$
* $g(x)=y\implies g'(x)=\frac{dy}{dx}$

$\therefore\frac{dy}{dx}*\cos(y)=1$

$\frac{dy}{dx}=\frac{1}{\cos(y)}$

Given that: $y=\arcsin(x)$

$\therefore\frac{dy}{dx}=\frac{1}{\cos(\arcsin(x))}$

According to trigonometry: $\sin(x)^2+\cos(x)^2=1$

Given that:
* $y=\arcsin(x)=\theta=$ angle of the Right triangle
* $\sin(y)=x$
* $\therefore x^2+\cos(y)^2=1$
* $\cos(y)^2=1-x^2$
* $\cos(\arcsin(x))=\cos(y)=(1-x^2)^{\frac{1}{2}}$

$\therefore\frac{dy}{dx}=\frac{1}{(1-x^2)^{\frac{1}{2}}}$

$\therefore\frac{dy}{dx}=(1-x^2)^{\frac{-1}{2}}$

**Find: $\frac{d}{dx}[\arccos(x)]$**

Given that: $y=\arccos(x)$

$\therefore \cos(y)=x$

$\frac{dy}{dx}[\cos(y)]=\frac{dy}{dx}[x]$

$\frac{dy}{dx}[\cos(y)]=1$

According to Chain Rule: $\frac{d}{dx}[f(g(x))] = (f\circ g)'= f'(g(x))*g'(x)$

According to Trigometry: $\frac{d}{dx}[\cos(x)]=-\sin(x)$

Given that:
* $f(x)=\cos(x)\implies f'(x)=-\sin(x)$
* $g(x)=y\implies g'(x)=\frac{dy}{dx}$

$\therefore-\frac{dy}{dx}*\sin(y)=1$

$\frac{dy}{dx}=\frac{-1}{\sin(y)}$

Given that: $y=\arccos(x)$

$\therefore\frac{dy}{dx}=\frac{-1}{\sin(\arccos(x))}$

According to trigonometry: $\sin(x)^2+\cos(x)^2=1$

Given that:
* $y=\arccos(x)=\theta=$ angle of the Right triangle
* $\cos(y)=x$
* $\therefore x^2+\sin(y)^2=1$
* $\sin(y)^2=1-x^2$
* $\sin(\arccos(x))=\sin(y)=(1-x^2)^{\frac{1}{2}}$

$\therefore\frac{dy}{dx}=\frac{-1}{(1-x^2)^{\frac{1}{2}}}$

$\therefore\frac{dy}{dx}=-(1-x^2)^{\frac{-1}{2}}$

**Find: $\frac{d}{dx}[\arctan(x)]$**

Given that: $y=\arctan(x)$

$\therefore \tan(y)=x$

$\frac{dy}{dx}[\tan(y)]=\frac{dy}{dx}[x]$

$\frac{dy}{dx}[\tan(y)]=1$

According to Chain Rule: $\frac{d}{dx}[f(g(x))] = (f\circ g)'= f'(g(x))*g'(x)$

According to Trigometry: $\frac{d}{dx}[\tan(x)]=\frac{1}{\cos(x)^2}$

Given that:
* $f(x)=\tan(x)\implies f'(x)=\frac{1}{\cos(x)^2}$
* $g(x)=y\implies g'(x)=\frac{dy}{dx}$

$\therefore\frac{dy}{dx}*\frac{1}{\cos(y)^2}=1$

$\frac{dy}{dx}=\cos(y)^2$

Given that: $y=\arctan(x)$

$\therefore\frac{dy}{dx}=\cos(\arctan(x))^2$

**Find: $\cos(\arctan(x))^2$**

Given that:
* $\frac{dy}{dx}=\cos(\arctan(x))^2$
* $y=\arctan(x)=\theta=$ angle of the Right triangle
* $\sin(\arctan(x))=\frac{b}{c}$
* $\cos(\arctan(x))=\frac{a}{c}$
* $\tan(\arctan(x))=x$

$\tan(\arctan(x))=\frac{\frac{b}{c}}{\frac{a}{c}}=\frac{b}{a}$

$\therefore a=1, b=x$

According to Pythagorean theorem: $a^2+b^2=c^2$

$\therefore 1+x^2=c^2\implies c=(1+x^2)^{\frac{1}{2}}$

Given that:
* $a=1$
* $b=x$
* $c=(1+x^2)^{\frac{1}{2}}$

Therefore:
* $\sin(\arctan(x))=\frac{b}{c}=x*(1+x^2)^{\frac{-1}{2}}$
* $\cos(\arctan(x))=\frac{a}{c}=(1+x^2)^{\frac{-1}{2}}$

$\therefore\cos(\arctan(x))^2=((1+x^2)^{\frac{-1}{2}})^2$

$\cos(\arctan(x))^2=(1+x^2)^{-1}$

$\cos(\arctan(x))^2=\frac{1}{(1+x^2)}$

$\therefore\frac{d}{dx}[\arctan(x)]=(1+x^2)^{-1}$
