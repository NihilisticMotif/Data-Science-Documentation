###(1) Inverse Function

**What is an inverse function?**

Inverse Function is the function that undo the original function

$f^{-1}(f(x))=x$

**Example No.1: $f^{-1}(x)$ of $f(x)=\frac{5-x}{3x}$**

Given that:
* $y=\frac{5-x}{3x}$

$\therefore y=\frac{5}{3x}-\frac{x}{3x}$

$y=\frac{5}{3x}-\frac{1}{3}$

$y+\frac{1}{3}-\frac{5}{3x}=0$

$-\frac{5}{3x}=-(y+\frac{1}{3})$

$\frac{5}{3x}=y+\frac{1}{3}$

$\frac{1}{x}=\frac{3}{5}*(y+\frac{1}{3})$

$x=\frac{1}{\frac{3}{5}*(y+\frac{1}{3})}$

$x=\frac{5}{3(y+\frac{1}{3})}$

$x=\frac{5}{3y+1}$

$\therefore f^{-1}(x)=\frac{5}{3y+1}$

**Property of a function**
* A function each input ($x$) has only one output ($y,f(x)$)

**Property of an inverse function**
* $f^{-1}(f(x))=f(f^{-1}(x))=x$
* $f^{-1}(x)\neq\frac{1}{f(x)}$
* Only function that have one input value ($x$) for every output value ($y$ or $f(x)$), have an inverse function.
* The domain (possible input value) of $f(x)$ is same as the range (possible output value) of $f^{-1}$ and vice versa.

###(-2) Binomial theorem: $(x+y)^n=\Sigma_{r=0}^n\frac{n!}{(n-r)!*r!}x^{n-r}y^r$

Notation Note:

* $\frac{n!}{(n-r)!r!}=\frac{n*(n-1)*...*(n-r+1)}{r!}=\begin{pmatrix}
n\\
r
\end{pmatrix}$

**Proof No.1**

Given that:
* $x,y\in\mathbb{R}$
* $n\in\mathbb{Z},n\geq2$
* $(x+y)^2=x^2+2xy+y^2$

Assume that:
* $(x+y)^n=\Sigma_{r=0}^n\frac{n!}{(n-r)!*r!}x^{n-r}y^r$

$(x+y)^n=\frac{n!}{(n-0)!*0!}x^{n-0}y^0+\frac{n!}{(n-1)!*1!}x^{n-1}y^1+...+\frac{n!}{(n-(n-1))!*(n-1)!}x^{n-(n-1)}y^{n-1}+\frac{n!}{(n-n)!*n!}x^{n-n}y^n$

$(x+y)^n=x^n+nx^{n-1}y+...+nxy^{n-1}+y^n$

if:
* $n=k+1$

$(x+y)^{n}=(x+y)*(x+y)^k$

$(x+y)^n=(x+y)*\Sigma_{r=0}^k\frac{k!}{(k-r)!*r!}x^{k-r}y^r$

$(x+y)^n=(x+y)*(x^k+kx^{k-1}y+...+kxy^{k-1}+y^k)$

$(x+y)^n=(x+y)x^k+(x+y)kx^{k-1}y+...+(x+y)kxy^{k-1}+(x+y)y^k$

$(x+y)^n=xx^k+yx^k+xkx^{k-1}y+ykx^{k-1}y+...+xkxy^{k-1}+ykxy^{k-1}+xy^k+yy^k$

$(x+y)^n=x^{k+1}+x^ky+kx^ky+kx^{k-1}y^2+...+kx^2y^{k-1}+kxy^k+xy^k+y^{k+1}$

$(x+y)^n=x^{k+1}+(k+1)x^ky+kx^{k-1}y^2+...+kx^2y^{k-1}+(k+1)xy^k+y^{k+1}$

$(x+y)^n=x^{k+1}+(k+1)x^ky+...+(k+1)xy^k+y^{k+1}$

$\therefore (x+y)^n=\Sigma_{r=0}^{k+1}\frac{(k+1)!}{((k+1)-r)!*r!}x^{(k+1)-r}y^r$
