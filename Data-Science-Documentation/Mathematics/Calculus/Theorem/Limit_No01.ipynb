###(1) Introduction of Limit

Other Keyword
* polynomial = an expression that consists of a sum of terms containing integer powers of $x$ 
* e.g. $3x^3-6x^2+17x-188$
* Rational Expressions = a fraction whose numerator and denominator are polynomials. 
* e.g. $\frac{2x^6-16x^2+17x-188}{3x^3-6x^2+17x-188}$, $\frac{x^0+17}{3x^3-6x^2+17x-188}$
* Quotient = a result obtained by dividing one quantity by another.
* Domain = the set of all possible inputs for the function. 

Limits = the value that a function (or sequence) approaches as the input (or index) approaches some value.

Limit Notation

$\lim_{x \to c}f(x) = a$

Output value of function $f(x)$ as input value $x$ approaches to value $c$ is equal to value $a$

$\lim_{x \to c^+}g(x) = a$

Output value of function $g(x)$ as input value $x$ approaches to value $c$ is equal to value $a$ from right to left size of x axis.

$\lim_{x \to c^-}g(x) = b$

Output value of function $g(x)$ as input value $x$ approaches to value $c$ is equal to value $b$ from left to right size of x axis.

if $\lim_{x \to c^+}g(x) \neq \lim_{x \to c^-}g(x)$
$\therefore \lim_{x \to c}g(x)$ is undifined.

if $\lim_{x \to c^+}g(x) = \lim_{x \to c^-}g(x)$
$\therefore \lim_{x \to c}g(x)$ is difined.

Property of Limit
1. When a limit doesn't approach the same value from both sides, we say that the limit doesn't exist.

###(2) Limit Rules, $+,-,x,\div$ of ($\lim \neq \infty$)

1. Multiply by Constant (Constant Property)

$\lim_{x \to a}[f(x)*c] = c*\lim_{x \to a}f(x) = c*F$

2. Sum and Difference Property

$\lim_{x \to a}[f(x)\pm g(x)] = \lim_{x \to a}f(x) \pm \lim_{x \to a}g(x) = F\pm G$

3. Multiplication (Product Property)

$\lim_{x \to a}[f(x)*g(x)] = \lim_{x \to a}f(x)*\lim_{x \to a}g(x)=F*G$

4. Division

$\lim_{x \to a}[\frac{f(x)}{g(x)}] = \frac{\lim_{x \to a}f(x)}{\lim_{x \to a}g(x)} = \frac{F}{G}$ for $G \neq 0$

5. Power by Constant

###(4) The Squeeze Theorem, $h(x) \leq g(x) \leq f(x)$ and $\lim_{x \to a}h(x) = \lim_{x \to a}f(x) = L$ then $\lim_{x \to a}g(x) =L$


$h(x) \leq g(x) \leq f(x)$ when $x$ is nearly $a$ and $\lim_{x \to a}h(x) = \lim_{x \to a}f(x) = L$ then $\lim_{x \to a}g(x) =L$

Example 1:

$h(x) = -x^2$ (blue line in Graph No.1)

$g(x) = g(x) $ (green line in Graph No.1)

$f(x) = x^2$ (red line in Graph No.1)

$\lim_{x \to 0}h(x) = \lim_{x \to 0}f(x) = 0$

$\therefore \lim_{x \to 0}g(x)=0$

```python

# importing the required module
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.linspace(-10,10,1000)


# corresponding y axis values
y1 = np.power(x,2)
y2 = np.power(x,2)*np.sin(1/x)
y3 = -np.power(x,2)
# plotting the points 
plt.plot(x, y1,'r')
plt.plot(x, y2,'g')
plt.plot(x, y3,'b')
  
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
  
# giving a title to my graph
plt.title('Graph No.1')
plt.grid()
# function to show the plot
plt.show()
```

###(6) Continuity

$f(x)$  is said to be continuous at $x=a$ if 
1. $\lim_{x \to a}f(x) = f(a)$
2. $\lim_{x \to a}f(x)$ and $f(a)$ are exist.

$f(x)$  is said to be continuous on the interval $(b,c)$ if $f(x)$  is said to be continuous at $x=a$ for every $a$ in $(b,c)$

$f(x)$  is said to be continuous on the interval $[b,c]$ if $f(x)$  is said to be continuous at $x=a$ for every $a$ in $[b,c]$


Type of Discontinuity
1. Jump Discontinuity
2. Removeable Discontinuity
3. Infinite Discontinuity
4. Wild Discontinuity

```python
# importing the required module
import matplotlib.pyplot as plt
import numpy as np
import math
from pylab import *

x=linspace(0,1,100)    
y=zeros(100)
y[x<0.5] = x[x<0.5]
y[x>=0.5] = 1 + x[x>=0.5]

pos = np.where(np.abs(np.diff(y)) >= 0.5)[0]+1
x = np.insert(x, pos, np.nan)
y = np.insert(y, pos, np.nan)

plot(x, y)

plt.title('''Jump Discontinuity
$f(x)${${g(x):x<0.5}$, ${h(x):x \geq 0.5}$}''')
plt.grid()
# function to show the plot
plt.show()
```
