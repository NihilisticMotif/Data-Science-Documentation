###(3) Inverse Function of Trigometry function

Note:
* We have to choose any domain and range of a function that satisfy the property of an inverse function

**Function No.1**
* Function: $y=\sin(x)$
* Domain of a function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Range of a function: $[-1,1]$
* Inverse Function: $y=\arcsin(x)$
* Domain of an inverse function: $[-1,1]$
* Range of an inverse function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Note: $\arcsin(x)\neq\frac{1}{\sin(x)}$

**Function No.2**
* Function: $y=\cos(x)$
* Domain of a function: $[0,\pi]$
* Range of a function: $[-1,1]$
* Inverse Function: $y=\arccos(x)$
* Domain of an inverse function: $[-1,1]$
* Range of an inverse function: $[0,\pi]$
* Note: $\arccos(x)\neq\frac{1}{\cos(x)}$

**Function No.1**
* Function: $y=\tan(x)$
* Domain of a function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Range of a function: $[-\infty,\infty]$
* Inverse Function: $y=\arctan(x)$
* Domain of an inverse function: $[-\infty,\infty]$
* Range of an inverse function: $[-\frac{\pi}{2},\frac{\pi}{2}]$
* Note: $\arctan(x)\neq\frac{1}{\tan(x)}$

#Formula

###Property of Trigonometry function
* $\sin(x+2\pi*c)=\sin(x)$ if $c$ is integer.
* $\cos(x+2\pi*c)=\cos(x)$ if $c$ is integer.
* $\sin(-x)=-\sin(x)$
* $\cos(-x)=\cos(x)$
* $\tan(-x)=-\tan(x)$
* $\sin(x)^2+\cos(x)^2=1$
* $\sin(x) =\cos(\frac{\pi}{2}-x)$
* $\cos(x) =\sin(\frac{\pi}{2}-x)$

###Sum and Difference Radius Formulas

* $\sin(x+y) = \sin(x)\cos(y)+\cos(x)\sin(y)$

* $\sin(x-y) = \sin(x)\cos(y)+\cos(x)\sin(y)$

* $\cos(x+y) = \cos(x)\cos(y)–\sin(x)\sin(y)$

* $\cos(x-y) = \cos(x)\cos(y)+\sin(x)\sin(y)$

* $\tan(x+y) = \frac{(\tan(x)+\tan(y))}{(1 - \tan(x)\tan(y)}$
